```python
import pandas as pd
import numpy as np
```

```python
sharky=pd.read_csv('GSAF5.csv', encoding= 'iso-8859-1')
```

```python
sharky.head(10)
```

```python
#Revisamos un poco como es el dataframe
sharky.info()
```

```python
#Hay dos columnas que apenas tienen datos (Unnamed: 22 & Unnamed: 23).
#Lo confirmamos revisando el numero de nulls (solo para mostrar el código)
null_cols=sharky.isnull().sum()
null_cols  #[null_cols>0] La lista es corta, sino podriamos filtrar por los que tienen mas que 0 nulls
```

```python
#Revisamos las columnas que tienen bastantes nulls (>1000) para entender como funcionan
sharky[list (null_cols[null_cols>1000].index)].head(50)
```

```python
#Revisamos los valores no null de Unnamed: 22 & Unnamed: 23 para ver que son:
unnamed = list (null_cols[null_cols>5000].index)
#unnamed_df = sharky[(sharky['Unnamed: 22'].isna()== False) | (sharky['Unnamed: 23'].isna()== False)]
unnamed_df = sharky.dropna(subset=unnamed, how='all')
unnamed_df.head()
```

```python
#Como vemos que los valores para Unnamed 22 & 23 no tienen sentido borramos ambas columnas
sharky = sharky.drop(unnamed, axis=1)
sharky.head(50)
```

```python
#Actualmente tenemos 5992 filas. Borramos duplicados y comprobamos si se ha borrado alguno.
sharky = sharky.drop_duplicates()
len(sharky)
```

```python
#Listamos las columnas que tenemos ahora:
sharky.info()
```

```python
#Las columnas case number, case number.1 y case number.2 se parecen mucho. Lo comprobamos
compare1 = np.where((sharky['Case Number'] == sharky['Case Number.1']), True, False)
np.count_nonzero(compare1) #Contamos los que no son 0, es decir, los que no son False
```

```python
#Repetimos con la columna Case Number.2
compare2 = np.where((sharky['Case Number'] == sharky['Case Number.2']), True, False)
np.count_nonzero(compare2)
```

```python
#Como la mayoria (5979 y 5990 de 5992) de valores son iguales podemos borrar las columnas repetidas.
sharky = sharky.drop(['Case Number.1', 'Case Number.2'], axis=1)
sharky.head()
```

```python
#Comprobamos tambien con las columnas href formula y href
compare3 = np.where((sharky['href'] == sharky['href formula']), True, False)
np.count_nonzero(compare3)
```

```python
#Borramos la columna repetida.
sharky = sharky.drop('href formula', axis=1)
sharky.head()
```
